```
from textblob import TextBlob
import pandas as pd
```

```
df=pd.read_csv('translate.csv' , sep='delimiter',  engine='python',encoding='cp1254') #İngilizceye çevirilen veri seti yüklenmiştir.
```

```
sentimentResults = []

for str in df.values: 
  sentence = str[0]
  blob = TextBlob(sentence) #TextBlob kütüphanesi kullanılarak veri setinde bulunan cümleler blob nesnesine aktarılmıştır.
  row = []
  polarity = blob.sentiment.polarity #Cümlelerin duygu analizini gerçekleştirmek için sentiment ve polarity bileşenleri kullanılmıştır.

  #TextBlob'ta cümleler negatif ise -1, nötr ise 0 ve pozitif ise 1 ile ifade edilmektedir.
  #Çıkan duygu analizi skorlarındaki değerlendirmede -0,1 ve 0,1 arasındaki skorlar nötr olarak değerlendirilmiştir.
  #0,1'den büyük skorlara sahip cümleler pozitif, -0,1'den küçük skorlara sahip cümleler negatif olarak değerlendirilmiştir.

  print("Sentence:",sentence)
  if (polarity >= (-0.1)) and (polarity <= 0.1):
    print("neutral:",polarity)
    polarity = 0
  elif polarity > 0.1:
    print("positive",polarity)
    polarity = 1
  else: 
    print("negative",polarity)
    polarity = -1
  
  row.append(sentence)
  row.append(polarity)

  sentimentResults.append(row)
  
#Çıkan sonuçlar, excel dosyasına Sentence ve Polarity olarak iki sütun olarak aktarılmıştır.

excelData = pd.DataFrame(sentimentResults, columns=['Sentence','Polarity']) 

excelData.to_excel('TextBlob_SA.xlsx')
```

```
excelData['Polarity'].value_counts() #3614 nötr, 974 pozitif ve 413 negatif cümle elde edilmiştir.
```
